```python
#!/usr/bin/python
import pandas as pd
import numpy as np
 
df = pd.read_csv('../data/nytimes.csv')
```

```python
group_cols  = ['Age', 'Gender', 'Signed_In']
```

```python
all_cols = group_cols + ['Clicks', 'Impressions']
```

```python
dfg = df[all_cols].groupby(group_cols).agg([np.mean])
```

```python
dfg
```

```python
dfg.index
```

```python
dfg.columns
```

```python
dfg['Click_Thru_Mean'] = dfg['Clicks'] / dfg['Impressions']
```

```python
dfg['Click_Thru_Mean']
```

```python
dfx = dfg.drop(['Clicks', 'Impressions'], axis=1)
```

```python
dfx
```

```python
dfx.columns = ['Click_Thru_Mean']
```

```python
dfx
```

```python
dfx.to_csv('nytimes_aggregation.csv')
```
